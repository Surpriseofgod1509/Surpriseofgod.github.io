```python
# Qudus Ishola
# 29-Jan-2026
# ICE1
# ChatGBT
# -----------------------------------------
# Movie Budget Analyzer 
# -----------------------------------------
# Starting list of movies stored as (name, budget) tuples
movies = [
    ("Eternal Sunshine of the Spotless Mind", 20000000),
    ("Memento", 9000000),
    ("Requiem for a Dream", 4500000),
    ("Pirates of the Caribbean: On Stranger Tides", 379000000),
    ("Avengers: Age of Ultron", 365000000),
    ("Avengers: Endgame", 356000000),
    ("Incredibles 2", 200000000)
]

# Variables for the calculations

total_budget = 0
over_average_count = 0

```

```python
# Asking user to enter How Many Movies to watch
num_new_movies = int(input("Enter how many new movies you wish to add: "))
```

```python
#Input Loop to Collect New Movies
for i in range(num_new_movies):
    name = input("Enter new movie name: ")
     # ask user to put the budget 
    budget = int(input("Enter new movie budget: "))
    movies.append((name, budget))
```

```python
#  Calculation
for movie in movies:
    total_budget += movie[1]

average_budget = total_budget // len(movies)
print("\nThe average budget of the movies is:", average_budget)
```

```python

# Phase Comparison
for movie in movies:
    movie_name = movie[0]
    movie_budget = movie[1]

    if movie_budget > average_budget:
        over_amount = movie_budget - average_budget
        print(
            movie_name,
            "cost $",
            movie_budget,
            ": $",
            over_amount,
            "over average.",
            sep=""
        )
        over_average_count += 1
```

```python
# Final Summary
print("\nThere were", over_average_count, "movies with over average budgets.")
```
